```python
import csv

f = open('incheon_weather.csv','r')
data = csv.reader(f)
header = next(data)

for row in data:
    print(row[-1]) # 먼저 print 함으로써 어느 위치에서 문제가 생겼는지 파악가능
    row[-1] = float(row[4]) # ' ' 데이터 값이 공란일 경우 변환 불가능

    
f.close()
```
