# 매년 내 생일때의 최고 최저 기온 데이터 시각화하기
## 201601555 강의진
* 1997년 7월 12일 ~ 현재
* 마이너스 부호 표시
    * plt.rcParams

```python
import csv

f = open('incheon_weather.csv','r')
data = csv.reader(f)
header = next(data) # header 사용 안함
max_temp = []
min_temp = []
# print(header)
for row in data:
    if row[-1] !='' and row[-2] !='':
        if int(row[0].split('-')[0])>=1997:
            month = row[0].split('-')[1]
            day = row[0].split('-')[2]
            if month == '07' and day == '12':
                max_temp.append(float(row[-1]))
                min_temp.append(float(row[-2]))
f.close()
```

```python
import matplotlib.pyplot as plt

plt.rc('font', family='Malgun Gothic')
plt.title('매년 생일때의 최고 최저기온 그래프')

# plt.rcParams['axes.unicode_minus'] = False 마이너스표시
plt.plot(max_temp, color='r', label='최고 기온')
plt.plot(min_temp, color='b', ls='--', label='최저 기온')

plt.legend()
plt.show()
```
