# draw_graph
* import matplotlib.pyplot
    * plot() 직선 또는 꺾은선 그래프
* 그래프에 제목 넣기
    * plt.title('제목')
    * 한글일 경우 폰트 설정
* 범례 넣기
    * 두개 이상의 데이터를 그래프에 표시할 때
    * plt.legend() 그래프를 보여주기전에 legend() 함수 실행
    * plt.plot(str,label='')
    * 범례 위치설정
* 그래프 색상 설정
    * plt.plot(str, color='색상', label)
* 그래프 선 모양 바꾸기
    * plt.plot(str, color, linestyle(=ls)='', label)

```python
import matplotlib.pyplot as plt

plt.plot([10, 20, 30, 40]) # plt.plot(list) # y축에 입력되어 있음, x축에 0부터 1씩 증가
plt.show()
```

```python
# %matplotlib inline 그래프를 주피터노트북 내에서 보고싶을 경우 (그래프가 웹상이 아닌 따로 실행되어질 경우)
import matplotlib.pyplot as plt

plt.rc('font', family = 'Malgun Gothic') # 폰트 설정 한글일 경우 깨짐
plt.title('타이틀')
plt.plot([1, 2, 3, 4], [10, 20, 30, 40]) # plt.plot(list1, list2) x, y축
plt.show()
```

```python
import matplotlib.pyplot as plt

plt.rc('font', family = 'Malgun Gothic') 
plt.title('타이틀')
plt.plot([1, 2, 3, 4], [10, 20, 30, 40], color = 'r', linestyle= '-', label='범례1') # 색상 및 선 모양
plt.plot([20, 40, 10, 20], color='b', ls='--', label='범례2') # x 축 범위를 설정 안해줬기 때문에 0 1 2 3에 대입

plt.legend() # default 값은 자동으로 그래프에 따라 설정 현재 2 ex) plt.legend(loc=위치)
plt.show()
# 범례 위치
# [2, 9, 1]
# [6, 10, 5,7]
# [3, 8 ,4]
```
